```python
# load libraries
import pandas as pd
import numpy as np
import requests
import tweepy
import json
import time
import os
import datetime
import string
import matplotlib.pyplot as plt
import math
import seaborn as sns
```

```python
twitter_clean = pd.read_csv("twitter_clean.csv", encoding = "utf-8")
```

```python
db = twitter_clean.groupby('p1').filter(lambda x: len(x) >= 25)

db['p1'].value_counts().plot(kind = 'barh', figsize=(18, 10))
plt.title('Frequency of rated Dog Breed')
plt.xlabel('Count')
plt.ylabel('Dog Breed');
plt.size=(18,10);
plt.savefig("frequency_rated_dog_breed.png")
```

```python
df = twitter_clean[twitter_clean['rating'] < 40]
df['rating'].hist(figsize=(18, 10));
plt.title("Twitter Data Dog Rating");
plt.xlabel("Rating");
plt.ylabel("Frequency of Tweets");
plt.savefig("twitter_ratings.png")
```

```python
sns.lmplot(x="retweets", 
           y="favorites", 
           data=twitter_clean,
           height = 9,
           aspect=1.5);

plt.title('Favorite Tweet vs. Retweet');
plt.xlabel('Retweet Frequency');
plt.ylabel('Favorite Frequency');
plt.savefig("retweets_favourites.png")
```

# Introduction

Data Wrangling is the act of gathering, preparing and ultimately providing a tidy dataset of high quality. The goal of data wrangling is to provide or rather have a dataset that can be used for programmatic analysis of all kind. It can be argued that data wrangling represents the most essential part in the entire data analysis process. It also represents the most time-consuming module in the entire process. 

Three insights were taken from Twitter's Dog dataset. 

## Insight 1

The first visualisation illustrates that golden retriever are the most rated dog breeds among all other dog breeds. After that, Labrador's are the most rated dogs amongst all others, followed by Pembroke.

## Insight 2

The second visualisation shows that the ratings are positively skewed. The majority of ratings are in and around 10 to 25. Only very little ratings are below 10 and nearly no ratings are aboev 25.

## Insight 3

The third visualisation exemplifies that favourite tweets and retweets are positively correlated with each other. This shows that once there is a tweet that has a certain popularity level, it gets frequently retweeted. The visualisation also shows that the more favourite a tweet is, the more often it is retweeted.
